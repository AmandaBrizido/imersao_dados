```python
df.rename(columns={'work_year': 'ano', 'experience_level': 'senioridade', 'employment_type': 'contrato', 'job_title': 'cargo', 'salary': 'salario', 'salary_currency': 'moeda', 'salary_in_usd': 'usd', 'employee_residence': 'residencia', 'remote_ratio': 'remoto', 'company_location': 'empresa', 'company_size': 'tamanho_empresa'}, inplace=True)
```

```python
df.head()
```

```python
df["senioridade"].value_counts()
"""
EN - Entry-level (Nível de entrada)
MI - Mid-level (Nível médio)
SE - Senior-level (Nível sênior)
EX - Executive-level (Nível executivo)
""" 
```

```python
df["contrato"].value_counts()
"""
FT - Full-time (Tempo integral)
PT - Part-time (Meio período)
CT - Contract (Contrato)
FL - Freelance (Freelancer)
"""
```

```python
df["remoto"].value_counts()
"""
0 - Presencial
50 - Híbrido
100 - Remoto
"""
```

```python
df["tamanho_empresa"].value_counts()
"""
M - Médio
S - Small (Pequeno)
L - Large (Grande)
"""
```

```python
df["senioridade"] = df["senioridade"].replace({
    'EN': 'Nível de entrada',
    'MI': 'Nível médio',
    'SE': 'Nível sênior',
    'EX': 'Nível executivo'
})
```

```python
df["remoto"] = df["remoto"].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})
```

```python
df["tamanho_empresa"] = df["tamanho_empresa"].replace({
    'M': 'Médio',
    'S': 'Pequeno',
    'L': 'Grande'
})
```

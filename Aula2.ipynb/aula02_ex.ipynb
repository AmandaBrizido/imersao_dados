```python
import pandas as pd
import numpy as np
```

```python
df_vendas = pd.DataFrame({
    'produto': ["Camiseta", "Calça", "Tênis", "Jaqueta"],
    'preço': [50, 80, 120, np.nan], 
    'quantidade': [10, 5, np.nan, 2]
})
```

```python
df_vendas.head()
```

```python
df_vendas.isnull()
```

```python
df_vendas.isnull().sum()
```

```python
df_vendas['vendas_media'] = df_vendas['quantidade'].fillna(df_vendas['quantidade'].mean().round(2))
df_vendas['valor_preenchido'] = df_vendas['preço'].fillna(0)
df_vendas
```

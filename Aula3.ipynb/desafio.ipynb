```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
df.columns
```

```python
df.rename(columns={'work_year': 'ano', 'experience_level': 'senioridade', 
                   'employment_type': 'tipo_contrato', 
                   'job_title': 'cargo', 
                   'salary': 'salario', 
                   'salary_currency': 'moeda', 
                   'salary_in_usd': 'usd', 
                   'employee_residence': 'residencia', 
                   'remote_ratio': 'remoto', 
                   'company_location': 'local_empresa', 
                   'company_size': 'tamanho_empresa'}, inplace=True)
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})
```

```python
df.head()
```

```python
df_limpo = df.dropna()
df_limpo = df_limpo.assign(ano=df_limpo['ano'].astype('int64'))
df_limpo.head()
```

```python
datascientist_salary = df_limpo[df_limpo['cargo'] == 'Data Scientist']
mean_salary = datascientist_salary.groupby('residencia')['usd'].mean().sort_values(ascending=False).reset_index()
img = px.bar(mean_salary,
    x='residencia', 
    y='usd',
    color= 'usd',
    color_continuous_scale= 'Blues',
    title='Data Scientist: Média Salarial por País',
    labels={'usd': 'Média Salarial Anual', 'residencia': 'Países'})
img.show()
```
